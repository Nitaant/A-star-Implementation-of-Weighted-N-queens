```python
import copy
import numpy as np
import random

#Random Matrix generator
n = input("Enter your value: ")
n=int(n)
positions = [random.randint(0,n-1) for x in range(n)]
weights = [random.randint(1,9) for x in range(n)]

def get_attacking_pairs(positions):
    attacking_pairs = []
    
    for i in range(len(positions)):
        attack = 0
        for j in range(len(positions)):
            if i != j:
                if positions[i] == positions[j]:
                    attack += 1
                if abs(positions[i] - positions[j]) == abs(i-j):
                    attack += 1
        attacking_pairs.append(attack)
    return attacking_pairs




def get_heuristic_value(attacking_pairs, weights, possible_move, current_position, total_path_cost):
    heuristic_values = 0
    
    #attacking pair
    for i in range(len(weights)):
        heuristic_values += ((weights[i])**2)*attacking_pairs[i]
    
    #movement cost
    for i, q in enumerate(range(len(possible_move))):
        diff = possible_move[i] - current_position[i]
        if diff == 0:
            index = i
            break
    
    heuristic_values += possible_move[index] * weights[index]
    heuristic_values += total_path_cost
    return heuristic_values



def get_all_possible_moves(positions):
    n = len(positions)
    n -= 1
    possible_boards = []
    for i in range(len(positions)):
        traverse_column = [i for i in range(n+1)]
        for every_step in traverse_column:
            if every_step != positions[i]:
                one_move = copy.deepcopy(positions)
                one_move[i] = every_step
                possible_boards.append(one_move)
    return possible_boards





def print_board(positions):
    board = []
    for row in range(len(positions)):
        row = []
        for column in range(len(positions)):
            row.append(0)
        board.append(row)
    for i, j in enumerate(positions):
        board[j][i] = "Q" 
    for row in board:
        for column in row:
            print(column, end='  ')
        print("")
```

```python
from time import sleep
def solution(positions, weights, min_h_value = 1):
    total_path_cost=0
    print_board(positions)
    print("\n ----- \n")
    closed_list = []
    path_cost = 0
    board = []
    print("Iterating through Different Positions")
    while min_h_value:
        all_possible_moves = get_all_possible_moves(positions)
        all_h_values = []
        for one_possible_board in all_possible_moves:
            attacking_pairs = get_attacking_pairs(one_possible_board)
            h_value = get_heuristic_value(attacking_pairs,weights,one_possible_board,positions,total_path_cost)
            all_h_values.append(h_value)
        
        min_h_value = min(all_h_values)
        board_index = all_h_values.index(min_h_value)
        next_board_positions = all_possible_moves[board_index]
        
        while next_board_positions in closed_list:
            all_h_values.pop(board_index)
            all_possible_moves.pop(board_index)
            min_h_value = min(all_h_values)
            board_index = all_h_values.index(min_h_value)
            next_board_positions = all_possible_moves[board_index]
        
        closed_list.append(next_board_positions)
        positions = next_board_positions #finalboard
        print(positions)
    
    print("\nSolved Board:")
    print_board(positions)
    return positions
```

```python
import time
t1 = time.time()
final_position = solution(positions, weights)
t2 = time.time()
print("\nTime Taken: ")
print(t2-t1)
```
